#### 1. Find the average score
A program is needed to prompt the user operator for exam scores, calculate the average score and display the result to the screen. The average score is calculated as the sum of the scores divided by the number of scores.

## Input: user's exam scores
## Process: add all the given scores and divided by the total no. of scores
## Output: the average of the given scores

BEGIN
    SET TOTAL TO 0
    SET COUNTER TO 0
    SET CONTINUE TO 0
    SET AVERAGE TO 0
    LOOP CONTINUE = 0
        DISPLAY MESSAGE("Please enter exam scores, if there is no more, please enter -1 :")
        GET SCORE FROM USER
        if SCORE is -1
            SET CONTINUE TO 1
        ELSE
            ADD SCORE TO TOTAL
            INCREMENT COUNTER BY 1
        ENDIF
    ENDLOOP
    IF COUNTER >= 1
        SET AVERAGE TO TOTAL DIVIDED BY COUNTER
    ELSE
        DISPLAY MESSAGE ("You have not input any score")
    ENDIF
    DISPLAY AVERAGE 

END


Input: 
    10, 20, -1
Output: 15

Input: 
    -1
Output: You have not input any score
        0



#### 2. How much water runs off a roof in a rain storm?
To calculate the runoff from any given rainfall: Take the dimensions of the footprint of the roof and convert them to inches. (So, a 50' x 20' roof is 600" x 240"). Multiply the roof dimensions by the number of inches of rainfall. As an example, 600" x 240" x 1" = 144,000 cubic inches of water for an inch of rainfall. Divide that result by 231 to get the number of gallons (because 1 gallon = 231 cubic inches). (144,000/231 = 623.38).

## Input: Length and width of roof in feet and rainfall in inches
## Process: Calculate the area of the roof in inches by multiplying length and width and 144, ## calculate runoff in gallons by mutlitplying area of the roof and rainfall and divided the
## number by 231.
## Output: runoff in gallons over the roof.

BEGIN
    DISPLAY MESSAGE("Please enter the length of the roof in feet: ")
    GET LENGTH FROM USER
    DISPLAY MESSAGE("Please enter the width of the roof in feet: ")
    GET WIDTH FROM USER
    DISPLAY MESSAGE("Please enter the rainfall in inches: ")
    GET RAINFALL FROM USER
    SET RUNOFF TO LENGTH TIMES WIDTH TIMES 144 TIMES RAINFALL DIVIDED BY 231
    DISPLAY RUNOFF
END

Input:
    10, 20, 1
Output: 
    124.67 

Input:
    3, 4, 2
Output: 
    14.96

#### 3. Process customer record
A program is required to read a customer’s name, a purchase amount and a tax code. The tax code has been validated and will be one of the following: 0 tax exempt (0%) 1 state sales tax only (3%) 2 federal and state sales tax (5%) 3 special sales tax (7%) The program must then compute the sales tax and the total amount due,and print the customer’s name, purchase amount, sales tax and total amount due.

## Input: Customer's name, purchase amount and tax code
## Process: Calculate sales tax by multiplying purchase amount and the approprate tax rate ## depending on the tax code given. If tax code is 0, tax rate is 0; if tax code is 1, tax rate is ## .03; if tax code is 2, tax rate is .05; if tax rate is 3, tax code is .07. Calculate total amount ## by adding the purchase amount and the sales tax
## Output: Customer's name, purchase amount, sales tax and total amount.

BEGIN
    DISPLAY MESSAGE("Please enter customer's name: ")
    GET NAME FROM USER
    DISPLAY MESSAGE("Please enter purchase amount: ")
    GET PURCHASE_AMOUNT FROM USER
    DISPLAY MESSAGE("Please enter tax code: ")
    GET TAXCODE FROM USER
    IF TAXCODE is 0
        SET TAXRATE TO 0
    ELSE IF TAXCODE is 1
        SET TAXRATE TO .03
    ELSE IF TAXCODE is 2
        SET TAXRATE TO .05
    ELSE IF TAXCODE is 3
        SET TAXCODE TO .07
    ENDIF
    
    SET SALESTAX TO PURCHASE_AMOUNT MULTIPLY BY TAXCODE
    SET TOTALAMOUNT TO PURCHASE_AMOUNT PLUS SALESTAX
    DISPLAY NAME, PURCHASE_AMOUNT, SALESTAX, TOTALAMOUNT

END

#### 4. Calculate employee's pay 
A program is required by a company to read an employee’s number, pay rate and the number of hours worked in a week. The program is then to validate the pay rate field and the hours worked field and, if valid, compute the employee’s weekly pay and then print it and the input data.

Validation: According to the company’s rules, the maximum hours an employee can work per week is 60 hours, and the maximum hourly rate is $25.00 per hour. If the hours worked field or the hourly rate field is out of range, the input data and an appropriate message are to be printed and the employee’s weekly pay is not to be calculated.

Weekly pay calculation: Weekly pay is calculated as hours worked times pay rate. If more than 35 hours are worked, payment for the overtime hours worked is calculated at time-and-a-half.

## Input: employee's number, payrate, and number of hours worked in a week.
## Process: 
## 1. Check to see if the number of hours worked is greater than 60, if true, then print a error message and stop. If false, then calculate overtime hours and normal hours. Any hour exceeding 35 is overtime hours. Normal hour is any hour below or equals to 35.
## 2. Check to see if payrate is greater than 25, if true, then print a error message and stop. ## 3. Overtime pay is calculated by multiplying overtime hour by hourly rate by 1.5
## 4. Normal pay is calculated by multiply normal hour worked by hourly rate.
## 4. Calculate Weekly pay by adding up overtime pay and normal pay. 
## Output: Weekly pay and the input data -employee number, payrate, number of hours
## worked.

BEGIN

    DISPLAY MESSAGE("Please enter employee's number: ")
    GET EMPLOYEE_NUMBER FROM USER
    
    DISPLAY MESSAGE("Please enter payrate: ")
    GET PAYRATE FROM USER
    IF PAYRATE > 25
        DISPLAY PAYRATE
        DISPLAY MESSAGE ("Payrate cannot exceed 25")
    ELSE
        DISPLAY MESSAGE("Please enter number of hours worked last week: ")
        GET HOURS_WORKED FROM USER
        IF HOURS_WORKED > 60
            DISPLAY HOURS_WORKED
            DISPLAY MESSAGE ("Hours worked cannot exceed 60")
        ELSE
            SET OVERTIME_HOUR TO HOURS_WORKED -35
            IF OVERTIME_HOUR <0
                SET OVERTIME_HOUR TO 0
                SET NORMAL_HOUR TO HOURS_WORKED
            ELSE
                SET NORMAL_HOUR TO 35
            END IF
            
            SET NORMAL_PAY TO NORMAL_HOURS TIMES PAYRATE 
            SET OVERTIME_PAY TO OVERTIME_HOUR TIMES PAYRATE TIMES 1.5
            SET WEEKLY_PAY TO NORMAL_PAY PLUS OVERTIME_PAY
            DISPLAY WEEKLY_PAY, EMPLOYEE_NUMBER, PAYRATE, HOURS_WORKED
        ENDIF
    ENDIF
    
END
